## 4. Data preparation

```python
appstore = appstore.drop(['Unnamed: 0', 'id', 'url', 'icon_url', 'watch', 'ipad', 'ipod', 'iphone', 'tv', 'mac'], axis=1)
appstore
```

```python
nonull = appstore.dropna()
nonull
```

```python
nonull['release_date'] = pd.to_datetime(nonull['release_date'])
nonull['release_date'] = nonull['release_date'].dt.strftime('%Y')
sorted_years = nonull[['release_date']].sort_values(['release_date'], ascending=True)
sorted_years
```

```python
nofree = nonull.loc[nonull['price'] != 'Free'].loc[nonull['price'] != 'Size']
nofree['price'] = nofree['price'].str.replace('$','', regex=True).astype(float)
nofree[['price']]
```

## 5. Data analysis and visualisation

### 5.1 Q1: Show the number of games for each average rating
Description of your steps...

```python
# data preparation for Q1
nonull = appstore.dropna()
nonull.head()
```

```python
nonull['rating_avg'].value_counts()
```

```python
plt.figure(figsize = (16, 6))
sns.set_style("whitegrid")
sns.countplot(x='rating_avg', data = nonull)
plt.title('Average rating', size = 14)
```

### 5.2 Q2: Find the number of games released for each year and create a plot
Description of your steps...

```python
# data preparation for Q2
nonull['release_date'] = pd.to_datetime(nonull['release_date'])
nonull['release_date'] = nonull['release_date'].dt.strftime('%Y')
sorted_years = nonull[['release_date']].sort_values(['release_date'], ascending=True)
sorted_years.head()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x='release_date', data = sorted_years)
plt.title('Counts of games for each year', size = 14)
```

### 5.4 Q4: Find the percentage of each game genre in the entire dataset. Create a pie chart based on this data
Description of your steps...

```python
genresperc = nonull['sub_category'].value_counts()
```

```python
genresperc.plot.pie(autopct='%1.1f%%',figsize=(15,35))
plt.title('Pie plot', size = 14)
plt.show()
```

#### The App store has the most games in the Puzzle and Casual genres. Fewer games in the Music genre

### 5.5 Q5: Make up the top 5 most popular and unpopular genres among Not Free Games
Description of your steps...

```python
top5 = nofree["sub_category"].value_counts()[:5]
top5v2 = nofree["sub_category"].value_counts()[-5:]
nofree["sub_category"].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
top5.plot(kind="barh")
plt.xlabel('Count')
plt.title("Top 5 Most Popular Genres of Games", size=14)

plt.figure(figsize=(12,6))
top5v2.plot(kind="barh")
plt.xlabel('Count')
plt.title("Top 5 Unpopular Genres of Games", size=14)

plt.show()
```
